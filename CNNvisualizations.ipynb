```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers.convolutional import Convolution2D,MaxPooling2D
from keras.layers.core import Activation
```

```
cat = cv2.imread('cat.png')
```

```
plt.imshow(cat)
plt.grid(False)
```

```
cat.shape
```

```
model = Sequential()
model.add(Convolution2D(3,30,30,input_shape=cat.shape))
```

```
cat_batch = np.expand_dims(cat,axis=0)
cat_batch.shape
```

```
conv_cat = model.predict(cat_batch)
```

```
def visualize_cat(cat_batch):
  caty = np.squeeze(cat_batch,axis=0)
  print(caty.shape)
  catt = (caty - caty.min()) / (caty.max() - caty.min())
  plt.imshow(catt)
  plt.grid(False)
```

```
visualize_cat(conv_cat)
```

```
print("On increasing kernel size to 10x10")
visualize_cat(conv_cat)
```

```
print("On increasing kernel size to 30x30")
visualize_cat(conv_cat)
```

```
def nice_cat(model,cat):
  cat_batch = np.expand_dims(cat,axis=0)
  conv_cat2 = model.predict(cat_batch)
  print(conv_cat2.shape)
  conv_cat2 = np.squeeze(conv_cat2,axis=0)
  print(conv_cat2.shape)
  conv_cat2 = conv_cat2.reshape(conv_cat2.shape[:2])
  print(conv_cat2.shape)
  #catt = (conv_cat2 - conv_cat2.min()) / (conv_cat2.max() - conv_cat2.min())
  plt.imshow(conv_cat2)
  plt.grid(False)
```

```
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape = cat.shape))
```

```
nice_cat(model,cat)
```

```
model = Sequential()
model.add(Convolution2D(1,15,15,input_shape = cat.shape))
nice_cat(model,cat)
```

```
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape = cat.shape))
model.add(Activation('relu'))
nice_cat(model,cat)
```

```
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape = cat.shape))
model.add(MaxPooling2D(pool_size=(2,2)))
nice_cat(model,cat)
```

```
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape = cat.shape))
model.add(MaxPooling2D(pool_size=(5,5)))
nice_cat(model,cat)
```

```
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape = cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(5,5)))
nice_cat(model,cat)
```
